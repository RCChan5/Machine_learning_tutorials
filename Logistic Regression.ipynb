# Background
Before getting into logistic regression please refresh yourself on the Linear regression concepts.

Logistic regression is similar to linear regression except it predicts wether something is **True** or **False**, instead of predicting if something is continuous like size. Also instead of fitting a line to the data, Logistic regression fits an "S" shaped logistic function. 


If we weighed a heavy mouse there is a high probability that the mouse is obese. If we weighed a light mouse there is a high probability that the mouse is not obese

Although logistic regression tells the probability that a mouse is obese or not. It is usually used for classification

**Just like in linear regression we can make simple models:**

obesity is predicted by weight

**Or more complicatyed models**

Obesity is predicted by weight+Age+size+etc.

**Unlike normal regression** We can not copmare the complicated model to the simple model. Instead we test to see if a variables effect on the prediciton is significantly different from 0. 

We use **Wald's test** to figure this out refer to the walds test notes for more info

Over all logistics regressions ability to provide probabilities and classify new samples using continuous and discrete measurements make it popular for machine learning methods.

## Logistic regression vs Linear regression
One big difference between linear regression and logistic regression is how the line is fit to the data. With linear regresion we use least squares and use the residuals to calculate the R^2

Logistic regression doesnt have the same concept of a residual so it cant use least squares and it cant calculate r^2 instead it uses something called **Maximum Likelihood**
